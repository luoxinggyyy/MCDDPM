```python
import numpy as np
import torch
import pandas as pd
from plotnine import ggplot, aes, geom_line,geom_point,geom_abline,geom_smooth,labs,geom_density
```

```python
b = pd.read_csv('cc3.csv').dropna()
print(b.size)
```

```python
(
    ggplot(b, aes(x='real_por', y='test_por'))
    + geom_point()
    + geom_smooth()
    + geom_abline(intercept=45,
                  slope=0.1,
                  color='blue',      # set line colour
                  size=2,            # set line thickness
                #   linetype="dashed"  # set line type
                 )
    + labs(x='real_por', y='test_por')
)
```

```python
(
    ggplot(b, aes(x='dis_por',color = 'label',fill='label'))
    + geom_density(alpha = 0.1)
    
)
```
